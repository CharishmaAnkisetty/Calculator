```python
import time
import math

def sum(num1, num2):
    print("Addition:", num1 + num2)

def difference(num1, num2):
    print("Subtraction:", num1 - num2)

def product(num1, num2):
    print("Multiplication:", num1 * num2)

def division(num1, num2):
    if num2 == 0:
        print("Error: Division by zero!")
    else:
        print("Division:", num1 / num2)

def square(num):
    print("Square:", num ** 2)

def square_root(num):
    if num < 0:
        print("Error: Cannot calculate square root of a negative number!")
    else:
        print("Square Root:", math.sqrt(num))

while True:
    print("Select an operation to perform:")
    print("1. Adding")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Square")
    print("6. Square Root")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    if choice == '7':
        print("Calculator program exited.")
        print("-------------------")
        break

    if choice == '1':
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        sum(num1, num2)
        print("-------------------")
        time.sleep(1)
    elif choice == '2':
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        difference(num1, num2)
        print("-------------------")
        time.sleep(1)
    elif choice == '3':
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        product(num1, num2)
        print("-------------------")
        time.sleep(1)
    elif choice == '4':
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        division(num1, num2)
        print("-------------------")
        time.sleep(1)
    elif choice == '5':
        num = float(input("Enter a number: "))
        square(num)
        print("-------------------")
        time.sleep(1)
    elif choice == '6':
        num = float(input("Enter a number: "))
        square_root(num)
        print("-------------------")
        time.sleep(1)
    else:
        print("Input is out of range! Try 1-7")
        print("-------------------")
        time.sleep(1)
```
